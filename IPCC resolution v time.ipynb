```python
# List comprehensions to convert list to plotable data
yr = numpy.array( [ l['year'] for l in ipcc_models ] )
eres = numpy.array( [ l['eqres'] for l in ipcc_models ] )
mres = numpy.array( [ l['merres'] for l in ipcc_models ] )
zres = numpy.array( [ l['zonres'] for l in ipcc_models ] )
nlev = numpy.array( [ l['levels'] for l in ipcc_models ] )

# Equivalent horizontal resolution 
# (converts vertrical levels to zonal resolution and combines meridional/zonal/vertical into one length scale)
# deg2km approximately converts zonal resolutino to km
# mdeg2km average of cos(lat)
deg2km = 112.0
mdeg2km = 71.0
eff_km = ( (eres*zres)**0.5 * deg2km * mres * mdeg2km * (40*deg2km/nlev) )**0.33
```

```python
syr, mdn, mn, mini = [],[],[], []
for y in numpy.unique(yr):
    syr.append( numpy.int(numpy.median( yr[yr==y] )) )
    mdn.append( numpy.median( eff_km[yr==y] ) )
    mn.append( numpy.mean( eff_km[yr==y] ) )
    mini.append( numpy.min( eff_km[yr==y] ) )
ars = ['', 'FAR','SAR', 'TAR', 'AR4', 'AR5', 'AR6']
```

```python
# F-K's extrapolation
fyr = numpy.linspace(syr[0],2100,2) # years to plot fit over

k=slice(None,-1)
p = numpy.poly1d( numpy.polyfit(syr[k], numpy.log10(mini[k]), 1) )
leadn = 10**p(fyr) # Bleeding edge fit
print('Bleeding edge =',p,'Doubling time =',-numpy.log10(2)/p[1],'years')

p = numpy.poly1d( numpy.polyfit(syr[k], numpy.log10(mdn[k]), 1) )
medn = 10**p(fyr) # Median fit
print('Median fit =',p,'Doubling time =',-numpy.log10(2)/p[1],'years')

# Representation of Moore's law
mdb = 12/18 # Moore's doubling time in years
moore = mini[0]/( 2**((1/4)*(fyr-syr[0])*mdb) )

# Reproduce F-K's plot
plt.semilogy( yr, eff_km, 'b.') # individual models
plt.semilogy( syr[k], mdn[k], 'k-', label='Median') # median resolution
plt.semilogy( fyr, medn, 'k--', label='Median fit')
plt.semilogy( syr[k], mini[k], 'r-', label='Finest') # finest resolution
plt.semilogy( fyr, leadn, 'r--', label='Finest fit') # Bleeding edge
plt.semilogy( fyr, moore, 'b:', label="Moore's Law")
# Plot AR labels
for i in range(len(syr)):
    plt.text(syr[i], 800*(2-i%2), ars[i], horizontalalignment='center')
# Annotations
y = 230; plt.gca().axhline(y, color='k', linestyle=':')
plt.text(2025,y*1.2,'1st Rossby radius')
y = 10; plt.gca().axhline(y, color='k', linestyle=':')
plt.text(1985,y*1.2,'Mesoscale')
y = 0.1; plt.gca().axhline(y, color='k', linestyle=':')
plt.text(1985,y*1.2,'Sub-mesoscale')
y = 5e-3; plt.gca().axhline(y, color='k', linestyle=':')
plt.text(1985,y*1.2,'Langmuir turbulence')
plt.xlim(1980,2100); plt.xlabel('Year')
plt.ylim(1e-3,3e3); plt.ylabel('Equivalent Ocean Model Resolution (km)')
plt.legend()
plt.title('Resolution of Ocean Component of Coupled IPCC models')
plt.text(1981,1.2e-3,'(based on Fox-Kemper et al., 2014)', horizontalalignment='left', fontsize=8, color='grey');
```

```python
idof = 360/zres # DOF in zonal direction
jdof = 0.9 * 180/mres + .1* 180/eres # DOF in meridional direction
hdof = idof * jdof # Horizontal DOF
dof = ijdof * nlev
dx  = numpy.minimum( numpy.minimum( eres, zres) * deg2km, mres * mdeg2km )*1e3 # Smallest dx (in m)
dt = numpy.minimum( dx/3/4, 3600)
nt = numpy.round( 1 * 365 * 86400 / dt )
cost = dof*nt

cost_mdn, cost_max = [],[]
for y in numpy.unique(yr):
    cost_mdn.append( numpy.median( cost[yr==y] ) )
    cost_max.append( numpy.max( cost[yr==y] ) )
```

```python
k=slice(None,-1)
p = numpy.poly1d( numpy.polyfit(syr[k], numpy.log10(cost_max[k]), 1) )
cost_lead = 10**p(fyr) # Bleeding edge fit
print('Bleeding edge =',p,'Doubling time =',numpy.log10(2)/p[1],'years')

p = numpy.poly1d( numpy.polyfit(syr[k], numpy.log10(cost_mdn[k]), 1) )
cost_exp = 10**p(fyr) # Bleeding edge fit
print('Expected mdn =',p,'Doubling time =',numpy.log10(2)/p[1],'years')

plt.semilogy( yr, cost, '.')
plt.semilogy( syr[k], cost_mdn[k], 'k-', label='Median')
plt.semilogy( fyr, cost_exp, 'k--', label='Median fit')
plt.semilogy( syr[k], cost_max[k], 'r-', label='Biggest')
plt.semilogy( fyr, cost_lead, 'r--', label='Biggest fit')
plt.semilogy( fyr, 4e8 * 2**((fyr-1989)*(12/18)), 'b:', label="Moore's Law" )
plt.legend()
plt.xlabel('Year')
plt.ylabel('~DOF x steps/SMY')
plt.title('"Work" per simulated year')
```

```python
edx = zres*deg2km
e_mdn, e_min = [],[]
for y in numpy.unique(yr):
    e_mdn.append( numpy.median( edx[yr==y] ) )
    e_min.append( numpy.min( edx[yr==y] ) )

k=slice(None,-1)
p = numpy.poly1d( numpy.polyfit(syr[k], numpy.log10(e_min[k]), 1) )
e_lead = 10**p(fyr) # Bleeding edge fit
print('Bleeding edge =',p,'Doubling time =',-numpy.log10(2)/p[1],'years')

p = numpy.poly1d( numpy.polyfit(syr[k], numpy.log10(e_mdn[k]), 1) )
e_exp = 10**p(fyr) # Bleeding edge fit
print('Expected mdn =',p,'Doubling time =',-numpy.log10(2)/p[1],'years')

mdb = 12/18 # Moore's doubling time in years
moore4 = e_lead[0]/( 2**((1/4)*(fyr-syr[0])*mdb) )
moore3 = e_lead[0]/( 2**((1/3)*(fyr-syr[0])*mdb) )

plt.semilogy( yr, edx, 'b.') # individual models
plt.semilogy( syr[k], e_mdn[k], 'k-', label='Median')
plt.semilogy( fyr, e_exp, 'k--', label='Median fit') # median resolution
plt.semilogy( syr[k], e_min[k], 'r-', label='Finest')
plt.semilogy( fyr, e_lead, 'r--', label='Finest fit') # median resolution
plt.semilogy( fyr, moore4, 'b:', label="Moore's t$^4$ Law")
plt.semilogy( fyr, moore3, 'c:', label="Moore's t$^3$ Law")
plt.legend()
plt.xlim(1980,2060); plt.xlabel('Year')
plt.ylim(1e-3,3e3); plt.ylabel('Along-Equator Ocean Model Resolution (km)')
plt.title('Equatorial Zonal Resolution of Ocean (IPCC)')
# Annotations
y = 230; plt.gca().axhline(y, color='k', linestyle=':')
plt.text(2025,y*1.2,'1st Rossby radius')
y = 10; plt.gca().axhline(y, color='k', linestyle=':')
plt.text(1985,y*1.2,'Mesoscale')
y = 0.1; plt.gca().axhline(y, color='k', linestyle=':')
plt.text(1985,y*1.2,'Sub-mesoscale')
```
